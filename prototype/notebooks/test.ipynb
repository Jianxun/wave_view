```python
import plotly.graph_objects as go
import numpy as np
import pandas as pd # pandas is imported but not used, can be removed if not needed elsewhere

t = np.linspace(start=0, stop=np.pi*20, num=1000)

fig = go.Figure()

fig.add_trace(
    go.Scatter(x=t, y=np.sin(t), yaxis="y")
)

fig.add_trace(
    go.Scatter(x=t, y=np.cos(t), yaxis="y2")
)

# Setup vertically stacked y-axes
fig.update_layout(
    xaxis=dict(
        domain=[0, 1],
        rangeslider=dict(visible=True)
        # fixedrange will be controlled by buttons
    ),
    yaxis=dict(
        domain=[0, 0.45]

        # fixedrange defaults to False, allowing zoom
    ),
    yaxis2=dict(
        domain=[0.55, 1],
        anchor="x"
        # fixedrange defaults to False, allowing zoom
    ),
    height=600,
    
    # Default dragmode
    dragmode='zoom', # Initial mode can be Zoom XY

    # Add modebar buttons
    updatemenus=[
        dict(
            type="buttons",
            direction="right",
            x=1,
            y=1.15, # Adjusted y to ensure buttons are well-placed
            buttons=list([
                dict(
                    label="Zoom XY",
                    method="relayout",
                    args=[{
                        "dragmode": "zoom",
                        "xaxis.fixedrange": False,
                        "yaxis.fixedrange": False, # Explicitly ensure y-axes are zoomable
                        "yaxis2.fixedrange": False
                    }]
                ),
                dict(
                    label="Zoom Y",
                    method="relayout",
                    args=[{
                        "dragmode": "zoom",      # Use 'zoom' dragmode
                        "xaxis.fixedrange": True, # Fix X-axis for Y-only zoom
                        "yaxis.fixedrange": False,
                        "yaxis2.fixedrange": False
                    }]
                ),
                dict(
                    label="Zoom X",
                    method="relayout",
                    args=[{
                        "dragmode": "zoom",      # Use 'zoom' dragmode
                        "xaxis.fixedrange": False, # Fix X-axis for Y-only zoom
                        "yaxis.fixedrange": True,
                        "yaxis2.fixedrange": True
                    }]
                ),
            ]),
            showactive=True,
            pad=dict(t=10) # Added padding for better button appearance
        )
    ]
)

fig.show()
```

```python
import plotly.graph_objects as go
import numpy as np
import ipywidgets as widgets
from IPython.display import display, clear_output

# --- Initial Data & FigureWidget ---
t = np.linspace(start=0, stop=np.pi*20, num=1000)
fig = go.FigureWidget()

# Store y-axis configurations and their order
# Order is important for stacking: bottom-most first
active_yaxis_ids = ['y', 'y2'] # Initial axes
next_yaxis_number = 3 # To generate y3, y4, etc.

# --- Helper Function to Calculate and Apply Y-Axis Domains ---
def update_all_yaxis_domains():
    num_axes = len(active_yaxis_ids)
    if num_axes == 0:
        return

    spacing = 0.05 # Percentage spacing between axes
    total_spacing = (num_axes - 1) * spacing if num_axes > 1 else 0
    available_height_for_axes = 1.0 - total_spacing
    
    if available_height_for_axes <= 0: # Avoid division by zero or negative height
        available_height_for_axes = 0.01 
        # Or handle this as an error/limit on number of axes

    height_per_axis = available_height_for_axes / num_axes
    
    current_bottom = 0
    new_layout_updates = {}

    for i, axis_id_str in enumerate(active_yaxis_ids):
        # For Plotly, 'y' is yaxis, 'y2' is yaxis2, etc.
        plotly_axis_key = f"yaxis{'' if axis_id_str == 'y' else axis_id_str[1:]}"

        domain_start = current_bottom
        domain_end = current_bottom + height_per_axis
        
        if plotly_axis_key not in fig.layout:
             new_layout_updates[plotly_axis_key] = go.layout.YAxis(
                title=f"Y-Axis {axis_id_str.upper()}", 
                anchor="x" # Important for independent stacking
            )
        
        # Update domain directly on the existing or new yaxis object in layout
        if plotly_axis_key in new_layout_updates:
            new_layout_updates[plotly_axis_key].domain = [domain_start, domain_end]
            new_layout_updates[plotly_axis_key].title = f"Y-Axis {axis_id_str.upper()}" # Keep title updated
        elif fig.layout[plotly_axis_key]:
             # Create a new YAxis object if modifying directly is problematic
            # For simplicity, we'll assume direct modification or that fig.layout handles it.
            # More robust: fig.layout[plotly_axis_key] = go.layout.YAxis(**existing_attrs, domain=new_domain)
            # However, direct assignment to fig.layout[plotly_axis_key].domain should work with FigureWidget
             pass # Domain will be set in the batch_update from current fig.layout


        current_bottom = domain_end + (spacing if i < num_axes - 1 else 0)
    
    # Apply updates in batch
    with fig.batch_update():
        for i, axis_id_str in enumerate(active_yaxis_ids):
            plotly_axis_key = f"yaxis{'' if axis_id_str == 'y' else axis_id_str[1:]}"
            domain_start = (i * (height_per_axis + spacing)) if num_axes > 1 and i > 0 else i * height_per_axis
            if i > 0 and num_axes > 1 : domain_start = (i * height_per_axis) + i*spacing
            else: domain_start = i * height_per_axis


            domain_end = domain_start + height_per_axis
            
            # Ensure domains are within [0,1] and start < end
            domain_start = max(0, min(domain_start, 1))
            domain_end = max(0, min(domain_end, 1))
            if domain_end <= domain_start: domain_end = domain_start + 0.01 # minimal height

            if plotly_axis_key not in fig.layout:
                 fig.layout[plotly_axis_key] = go.layout.YAxis(title=f"Y-Axis {axis_id_str.upper()}", anchor="x")
            
            fig.layout[plotly_axis_key].domain = [domain_start, domain_end]
            fig.layout[plotly_axis_key].title = f"Y-Axis {axis_id_str.upper()}" # Keep title updated

# --- Initial Plot Setup ---
fig.add_trace(go.Scatter(x=t, y=np.sin(t), name="Trace on Y1", yaxis=active_yaxis_ids[0]))
fig.add_trace(go.Scatter(x=t, y=np.cos(t), name="Trace on Y2", yaxis=active_yaxis_ids[1]))

with fig.batch_update():
    fig.layout.xaxis=dict(domain=[0, 1], rangeslider=dict(visible=True))
    # Initial domains are set by update_all_yaxis_domains
    fig.layout.height=700
    fig.layout.title_text="Dynamically Add/Remove Y-Axes"
update_all_yaxis_domains()


# --- Widgets for UI ---
add_yaxis_button = widgets.Button(description="Add Y-Axis")
remove_yaxis_dropdown = widgets.Dropdown(description="Remove Axis:", options=active_yaxis_ids)
remove_yaxis_button = widgets.Button(description="Remove Selected Y-Axis")
output_area = widgets.Output() # For messages or debugging

# --- Callback for Adding Y-Axis ---
def on_add_yaxis_clicked(b):
    global next_yaxis_number
    with output_area:
        clear_output(wait=True)
        new_axis_id = f"y{next_yaxis_number}"
        if new_axis_id in active_yaxis_ids: # Should not happen if next_yaxis_number is managed well
            print(f"Error: Axis {new_axis_id} already exists.")
            return

        active_yaxis_ids.append(new_axis_id)
        
        # Add a placeholder trace to the new axis
        with fig.batch_data_update():
             fig.add_trace(go.Scatter(x=t, y=np.random.rand(len(t)), name=f"Trace on {new_axis_id.upper()}", yaxis=new_axis_id))

        update_all_yaxis_domains() # This will also create the layout object for the new axis
        
        remove_yaxis_dropdown.options = active_yaxis_ids # Update dropdown
        print(f"Added Y-Axis: {new_axis_id}. Current axes: {active_yaxis_ids}")
        next_yaxis_number += 1

add_yaxis_button.on_click(on_add_yaxis_clicked)

# --- Callback for Removing Y-Axis ---
def on_remove_yaxis_clicked(b):
    global next_yaxis_number # although not directly used here, it's part of the state
    with output_area:
        clear_output(wait=True)
        axis_to_remove = remove_yaxis_dropdown.value
        
        if not axis_to_remove:
            print("No axis selected to remove.")
            return
        if len(active_yaxis_ids) <= 1: # Keep at least one y-axis
            print("Cannot remove the last Y-axis.")
            return

        # Remove traces associated with this y-axis
        traces_to_keep = [trace for trace in fig.data if getattr(trace, 'yaxis', 'y') != axis_to_remove]
        
        with fig.batch_update(): # Batch data and layout changes
            fig.data = traces_to_keep
            
            # Remove y-axis from layout
            plotly_axis_key_to_remove = f"yaxis{'' if axis_to_remove == 'y' else axis_to_remove[1:]}"
            if plotly_axis_key_to_remove in fig.layout:
                del fig.layout[plotly_axis_key_to_remove]
            
            active_yaxis_ids.remove(axis_to_remove)
            update_all_yaxis_domains() # Recalculate domains for remaining axes

        remove_yaxis_dropdown.options = active_yaxis_ids
        if active_yaxis_ids: # Select a valid option if list is not empty
             remove_yaxis_dropdown.value = active_yaxis_ids[0]
        else: # Should not happen due to len check, but good practice
             remove_yaxis_dropdown.value = None

        print(f"Removed Y-Axis: {axis_to_remove}. Current axes: {active_yaxis_ids}")

remove_yaxis_button.on_click(on_remove_yaxis_clicked)


# --- Display UI ---
controls = widgets.VBox([
    widgets.HBox([add_yaxis_button]),
    widgets.HBox([remove_yaxis_dropdown, remove_yaxis_button]),
    output_area
])
display(controls)
display(fig)
```
